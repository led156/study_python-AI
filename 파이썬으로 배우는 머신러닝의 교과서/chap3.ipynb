### 3.1.3 3차 함수 f(x) = (x-2)x(x+2) 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
    return (x-2)*x*(x+2)
```

```python
print(f(1))
```

```python
print(f(np.array([1, 2, 3]))) # ndarray
```

### 3.1.4 그리는 범위를 결정하기

```python
x = np.arange(-3, 3.5, 0.5) # -3부터 3까지 범위, 0.5 간격
print(x)
```

```python
plt.plot(x, f(x))
plt.show()
```

### 3.1.6 그래프를 장식하기

```python
# 함수 정의
def f2(x, w):
    return (x - w) * x * (x + 2)

# x 정의
x = np.linspace(-3, 3, 100)

# 차트 묘사 (&string& -> 기울기체)
plt.plot(x, f2(x, 2), color='black', label='$w=2$')
plt.plot(x, f2(x, 1), color='cornflowerblue', label='$w=1$')
# x축 바탕으로 f2() 함수를 그린다. color 색, label 라벨(범례)
plt.legend(loc="upper left") # 범례 표시 위치, upper/lower, left/right
plt.ylim(-15, 15) # y축 범위
plt.title('$f_2(x)$') # 그래프 타이틀
plt.xlabel('$x$') # x축의 레이블(이름)
plt.ylabel('$y$') # y축의 레이블(이름)
plt.grid(True) # 격자 그리드
plt.show()
```

### 3.1.7 그래프를 여러 개 보여주기

```python
plt.figure(figsize=(10, 3)) # figure 지정 (전체 영역의 크기)
plt.subplots_adjust(wspace=0.5, hspace=0.5) # 그래프 간격 지정 (좌우 및 상하 간격)
for i in range(6):
    plt.subplot(2, 3, i+1) # 그래프 묘사 위치 지정
    plt.title(i+1)
    plt.plot(x, f2(x, i), 'k')
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```

## 3.2 3차원 그래프 그리기

### 3.2.1 이변수 함수
        = 두 변수의 함수

```python
import numpy as np
import matplotlib.pyplot as plt

def f3(x0, x1):
    r = 2*x0**2 + x1**2
    ans = r*np.exp(-r)
    return ans

xn = 9
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
        
print(x0)

print(np.round(y, 1))
```

### 3.2.2 수치를 색으로 표현하기: pcolor

```python
plt.figure(figsize=(3.5, 3))
plt.gray()
plt.pcolor(y) # 2차원 ndarray
plt.colorbar() # 컬러바를 나타냄
plt.show()
```

### 3.2.3 함수의 표면을 표시: surface

```python
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0, x1) # x0, x1(1차원 array) 좌표를 통해 xx0, xx1 생성(2차원의 array)

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1, 1, 1, projection='3d') # 3차원 차트 생성 -> 그래프 Id 반환.
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
# surface 표시. rstride, cstride : 가로 세로 선 개수(수가 적을수록 조밀), alpha : 투명도.
ax.set_zticks((0, 0.2)) # z축 눈금 조정. (숫자 겹침 방지)
ax.view_init(75, -95) # 3차원 그래프 방향 조정. argm1 : 상하 회전 각도(0:옆, 90:위), argm2 : 좌우 회전 각도(+:시계, -:반시계)
plt.show()

# xn(해상도)를 높이면 더 부드러움.
```

### 3.2.4 등고선으로 표시: contour

```python
xn = 50 # 해상도 50x50 설정.
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
        
xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(1, figsize=(4,4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') # 등고선 플롯 작성. 높이 단계 : 5 -> 반환.
cont.clabel(fmt='%3.2f', fontsize=8) # 등고선에 숫자 지정.
plt.xlabel('$x_0', fontsize=14)
plt.ylabel('$x_1', fontsize=14)
plt.show()
```
